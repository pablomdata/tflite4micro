```python
import pandas as pd
```

```python
df = pd.read_csv("./data/ford_train.tsv", sep="\t", header=None)
```

```python
df.head()
```

```python
X = df.iloc[:,1:]
```

```python
#X = X.iloc[:,200:250]
```

## K-Means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2, n_init=100, random_state=0)
```

```python
km.fit(X)
```

```python
y_pred = km.predict(X)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(0,500), X.iloc[0])
plt.plot(range(0,500), X.iloc[1])
```

```python
y = df.iloc[:,0].values
y = (y+1)/2
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y,y_pred)
```

```python
km.cluster_centers_
```

```python
plt.plot(range(500),km.cluster_centers_[0], label='Centro cluster 0')
plt.plot(range(500),km.cluster_centers_[1], label='Centro cluster 1')
plt.legend()
```
